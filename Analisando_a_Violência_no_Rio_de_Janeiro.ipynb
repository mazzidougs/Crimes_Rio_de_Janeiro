## Informações Estatísticas da Violência no Rio de Janeiro


```
#resumo estatístico
df.describe()
```

```
#roubo_veiculo
#furto_veiculos
#recuperacao_veiculos
#verificação da média destas colunas
rvei = df.roubo_veiculo.mean()
fvei = df.furto_veiculos.mean()
mrec = df.recuperacao_veiculos.mean()
print(rvei)
print(fvei)
print(mrec)
```

```
#calculando a porcentagem de carros recuperados em relação aos veiculos roubados e furtados
valor_roubov = df.roubo_veiculo.value_counts().sum()
valor_furtov = df.furto_veiculos.value_counts().sum()
valor_recv = df.recuperacao_veiculos.value_counts().sum()
porc = '%'

#Apenas metade dos veículos roubados e furtados são recuperados
print('{}{}'.format(valor_recv/(valor_furtov+valor_roubov)*100, porc))
```

```
#max e min de hom_doloso
max_homdoloso = df.hom_doloso.max()
min_homdoloso = df.hom_doloso.min()
print(max_homdoloso, min_homdoloso)
```
